```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
result_logs = '../../results/eta_tel_b/pos.csv'

results = pd.read_csv(result_logs)
```

```python
fig, axes = plt.subplots(2, 2, gridspec_kw={'wspace':0.3, 'hspace':0.4}, sharey='row')

for i, (filter_name, subframe) in enumerate(results.groupby('filter')):
    axes[0][i].boxplot(subframe['x'])
    axes[0][i].set_xticklabels([filter_name])
    axes[0][i].set_title('{:.2f} +- {:.2f}'.format(subframe['x'].mean(), 
                                                   subframe['x'].std()))
    axes[0][0].set_ylabel('x')
    axes[1][0].set_ylabel('y')
    
    axes[1][i].boxplot(subframe['y'])
    axes[1][i].set_xticklabels([filter_name])
    axes[1][i].set_title('{:.2f} +- {:.2f}'.format(subframe['y'].mean(), 
                                                   subframe['y'].std()))
```
